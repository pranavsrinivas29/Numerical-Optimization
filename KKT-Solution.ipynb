```python
import numpy as np

def solve(H, b, C, c):
    b = np.array(b).ravel()
    c = np.array(c).ravel()
    H = np.array(H)
    C = np.array(C)
    
    assert len(H.shape) == 2
    assert len(C.shape) == 2
    assert H.shape[0] == H.shape[1]
    
    n = H.shape[0]
    m = C.shape[0]
    assert m <= n
    assert len(b) == n
    assert len(c) == m
    
    kkt = np.zeros((n+m, n+m))
    kkt[:n, :n] = H
    kkt[n:,:n] = C
    kkt[:n,n:] = C.transpose()
    
    f = np.zeros(n+m)
    f[:n] = -b
    f[n:] = c
    
    x = np.linalg.solve(kkt, f)[:n]
    return x
```

```python
test(solve)
```
